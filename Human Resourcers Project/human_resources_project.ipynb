# CODING TASK #7. PANDAS WITH FUNCTIONS

```python
# Let's read a CSV file using Pandas as follows
hr_df = pd.read_csv('Human_Resources.csv')
hr_df
```

- **Let's assume the daily rate has increased by 10%**
- **Define a function that increases all clients networth by a fixed value of 10% (for simplicity sake)**
- **assume that net worth increased by 10%**
- **You can apply a function to the DataFrame**

```python
def increase_net_worth(daily_rate):
    return daily_rate * 1.10

hr_df.DailyRate = hr_df.DailyRate.apply(increase_net_worth)
hr_df
```

```python
hr_df
```

**PRACTICE OPPORTUNITY #7:**
- **Define a function that doubles the DailyRate and adds $100**

```python
def increase_daily_rate(daily_rate2):
    return daily_rate2 * 2 + 100

hr_df.DailyRate = hr_df.DailyRate.apply(increase_daily_rate)
hr_df
```

# CODING TASK #8: HEATMAP AND CORRELATION

- **Several features such as 'MonthlyIncome' and 'TotalWorkingYears' are tail heavy**
- **Job level is strongly correlated with total working hours**
- **Monthly income is strongly correlated with Job level**
- **Monthly income is strongly correlated with total working hours**
- **Age is strongly correlated with monthly income**

```python
plt.figure(figsize=(22, 10))

corr = hr_df.select_dtypes("number").corr()

sns.heatmap(corr, annot = True, cmap='viridis', fmt=".2f")

plt.title('Correlation Heatmap', fontsize=18)
plt.show()
```

```python
plt.figure(figsize=(20, 8))

corr = hr_df[["MonthlyIncome", "TotalWorkingYears", "JobLevel", "Age"]].corr()
sns.heatmap(corr, annot = True, cmap='icefire', fmt=".2f")

plt.title('Correlation Heatmap', fontsize=18)
plt.show()
```
